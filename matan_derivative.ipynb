### Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1

```python
import numpy as np
import pandas as pd
import scipy
import sympy as sym
import seaborn as sns
```

```python
# create a "symbol" called x
x = sym.Symbol('x')

#Define function
f = x**5+4*sym.sin(2*x)+sym.cos(3*x+3)

#Calculating Derivative
derivative_f1 = f.diff(x)
derivative_f2 = derivative_f1.diff(x)

print('Функция: ', f)
print('Первая производная: ', derivative_f1)
print('Вторая производная: ', derivative_f2)
```

### Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
# create a "symbol" called x
x = sym.Symbol('x')

#Define function
f = sym.sin(2*x+1)**5

#Calculating Derivative
derivative_f1 = f.diff(x)
derivative_f2 = derivative_f1.diff(x)

print('Функция: ', f)
print('Первая производная: ', derivative_f1)
print('Вторая производная: ', derivative_f2)
```

```python
def div_1(x):
    div_1_func = 10*sym.sin(2*x + 1)**4*sym.cos(2*x + 1)
    return div_1_func.evalf()
```

```python
def div_2(x):
    div_2_func = -20*sym.sin(2*x + 1)**5 + 80*sym.sin(2*x + 1)**3*sym.cos(2*x + 1)**2
    return div_2_func.evalf()
```

```python
div_1_culc = pd.DataFrame({'x':[], 'y':[]})

for x in np.linspace(-5, 5, num = 120):
    new_row = {'x':x, 'y':div_1(x)}
    div_1_culc = div_1_culc.append(new_row, ignore_index=True)
    
div_1_culc["y"] = div_1_culc["y"].astype(np.float64)
```

```python
div_2_culc = pd.DataFrame({'x':[], 'y':[]})

for x in np.linspace(-5, 5, num = 120):
    new_row = {'x':x, 'y':div_2(x)}
    div_2_culc = div_2_culc.append(new_row, ignore_index=True)
    
div_2_culc["y"] = div_2_culc["y"].astype(np.float64)
```

```python
sns.lineplot(data = div_1_culc, x = div_1_culc.x, y = div_1_culc.y)
```

```python
sns.lineplot(data = div_2_culc, x = div_2_culc.x, y = div_2_culc.y)
```
